## **Loading Sepsis Dataset**

```python
#import libraries
import pandas as pd
import numpy as np
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"

#Setting up pandas
pd.set_option('display.max_rows', 500)
```

```python
#Read CSV file to dataframe
sepsisDf = pd.read_csv("SepsisDataSet.csv",header=0)
sepsisDf.head()
```

```python
#datatypes of each column
dataTypes = pd.DataFrame(sepsisDf.dtypes,columns=['Dtype'])
dataTypes.reset_index(level=0,inplace=True)
dataTypes.rename(columns={'index':'Column'},inplace=True)
dataTypes.head()
```

```python
sepsisDf[sepsisDf.columns[3:7]] = sepsisDf.iloc[:,3:7].apply(pd.to_datetime,errors= 'coerce')
sepsisDf.head(10)
```

```python
pd.DataFrame(sepsisDf.dtypes,columns=['Dtype'])
```

```python
#Replace 'yes' with 1 and 'no' with 0 
sepsisDf.replace(('yes','no','Yes','No'), (1,0,1,0), inplace=True)
sepsisDf.head()
```

```python
#Replace F an M in Sex column with 1 and 2 
sepsisDf['Sex'] = sepsisDf['Sex'].replace(('F','M'),(1,2))
sepsisDf.head()
```

```python
pd.DataFrame(sepsisDf.dtypes,columns=['Dtype'])
```

```python

# summarize the number of rows with missing values for each column
for i in range(sepsisDf.shape[1]):
	# count number of rows with missing values
	n_miss = sepsisDf[[i]].isnull().sum()
	perc = n_miss / sepsisDf.shape[0] * 100
	print('> %d, Missing: %d (%.1f%%)' % (i, n_miss, perc))
```

```python
sepsisDf.shape[0]
```

```python
sepsisDf = sepsisDf.dropna(axis = 0,how='all')
sepsisDf.shape[0]
```
