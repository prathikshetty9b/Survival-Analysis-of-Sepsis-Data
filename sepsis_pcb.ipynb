```python
#import python libraries
import pandas as pd
```

```python
#read sepsis data
sepsisDf = pd.read_csv("SepsisDataSet.csv",header= 0)

#drop last 49 records
sepsisDf = sepsisDf.dropna(how='all')
```

```python
sepsisDf.describe()
```

```python
for col in sepsisDf.columns:
    print(col)
```

```python
columnsDf = pd.DataFrame(sepsisDf.columns)
columnsDf.to_csv("columns.csv")
```

```python
#Replace Sex M -> 1  F->2

#Male -> 1
sepsisDf['Sex'] = sepsisDf['Sex'].replace(to_replace='M', value=1)

#Female -> 2
sepsisDf['Sex'] = sepsisDf['Sex'].replace(to_replace='F', value=2)
```

```python
#Replace Yes and No with 1 and 0
toReplaceColumn = ['Pnemonia',	'UTI',	'Acute_abdominal_infection',	'meningitis',	'soft_tissue_infection',	'bone_joint_infection',	'wound_infection',	'CABSI',	'Endocarditis',	'implantable_device_infection',	'Gynaecological_infection',	'Deep_organ_abscess',	'others',	'Hyperthermia',	'Hypothermia',	'altered_mental_status',	'Tachycardia',	'Tachypnea',	'Leukocytosis',	'Leukopenia',	'Hyperglycemia', 'Blood_culture_positive',	'HIV',	'HBsAg',	'Diabetes',	'steroid',	'previous_hospitaliation',	'Hypertention',	'Asthma',	'COPD',	'Renal_disease']

for column in toReplaceColumn:
    if(sepsisDf[column].dtype == 'O'):
        sepsisDf[column] = sepsisDf[column].str.lower()
        sepsisDf[column] = sepsisDf[column].replace(to_replace='no',value=0)
        sepsisDf[column] = sepsisDf[column].replace(to_replace='yes',value=1)

sepsisDf.tail()
```

```python
dropColumns = ['Speciality','DOAHOSP','IDOAICU','DOICU','DODHOSP','VAR00001','Clinical_Diagnosis','expired_within_48hr','CRP','immunocompromised','PROGNOSIS']
standardizeDf = sepsisDf.drop(columns = dropColumns)
standardizeDf.head()
```

```python
standardizeDf.describe()
standardizeDf.isnull().sum()
```

```python
#dropping some essential columns (Note -> Will be included once the missing data is handled)
essentialColumns = ['GCS','Gly_Hb','ESR','DLC']
```

```python
standardizeDf = standardizeDf.drop(columns = essentialColumns)
standardizeDf.isnull().sum()
```

```python
standardizeDf = standardizeDf.dropna()
standardizeDf.isnull().sum()
```
